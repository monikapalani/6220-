```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Select the relevant numerical columns (e.g., Age and Cholesterol)
numerical_data = df[["Age", "Cholesterol"]]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)

# Perform PCA (Principal Component Analysis) to get the eigenvectors
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame for the principal components
pca_df = pd.DataFrame(pca_result, columns=["PC1", "PC2"])

# Add the 'Heart_Disease' information for coloring
pca_df['Heart_Disease'] = df['Heart Disease']

# Scatter plot for the eigenvectors
plt.figure(figsize=(10, 8))
sns.scatterplot(x="PC1", y="PC2", hue="Heart_Disease", palette="Set1", data=pca_df, s=100)
plt.title("Eigenvectors Scatter Plot: PCA of Age and Cholesterol")
plt.xlabel("Principal Component 1 (PC1)")
plt.ylabel("Principal Component 2 (PC2)")
plt.legend(title="Heart Disease")
plt.show()

# Print the eigenvectors and eigenvalues
print("Eigenvalues: ", pca.explained_variance_)
print("Eigenvectors: \n", pca.components_)
```

```python
cov_mat = np.cov(X_train_std.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
print('Eigen vals: {}'.format(eigen_vals))
```

```python
tot = sum(eigen_vals)
var_exp = [(i / tot) for i in sorted(eigen_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

plt.figure(figsize=(15,5))
plt.bar(range(num_features), var_exp, alpha=.5, align='center', label='Single variance')
plt.step(range(num_features), cum_var_exp, where='mid', label='Cumsum variance')
plt.ylabel('Factor of variance')
plt.xlabel('Main components')
plt.legend(loc='best')
plt.show()
```

```python
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(len(eigen_vals))]
eigen_pairs.sort(key = lambda k: k[0], reverse=True)
W = np.hstack((eigen_pairs[0][1][:, np.newaxis],
               eigen_pairs[1][1][:, np.newaxis]))
print('Matrix W: \n{}'.format(W))
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the dataset
url = "/content/Heart_Disease_Prediction.csv"
df = pd.read_csv(url)

# Preprocessing: Separate features and target
X = df.drop(columns=['Heart Disease'])  # Replace 'Heart_Disease' with your target column
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Keeping two components for the biplot
X_pca = pca.fit_transform(X_scaled)

# Biplot function
def biplot(score, coeff, labels=None):
    """
    Creates a biplot showing the principal component scores and the loadings.

    Parameters:
        score : PCA-transformed data (2D)
        coeff : Loadings (principal axes in feature space)
        labels : Feature names for loadings
    """
    xs = score[:, 0]
    ys = score[:, 1]
    n = coeff.shape[0]

    # Plot principal components
    plt.figure(figsize=(10, 7))
    plt.scatter(xs, ys, alpha=0.5)
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.title("PCA Biplot")

    # Plot arrows (feature loadings)
    for i in range(n):
        plt.arrow(0, 0, coeff[i, 0], coeff[i, 1], color='r', alpha=0.5, head_width=0.02, head_length=0.02)
        if labels is None:
            plt.text(coeff[i, 0]*1.15, coeff[i, 1]*1.15, f"Var{i+1}", color='g', ha='center', va='center')
        else:
            plt.text(coeff[i, 0]*1.15, coeff[i, 1]*1.15, labels[i], color='g', ha='center', va='center')

    plt.grid()
    plt.axhline(0, color='grey', lw=0.5)
    plt.axvline(0, color='grey', lw=0.5)
    plt.show()

# Get loadings (coefficients) and call biplot
pca_components = pca.components_.T  # Transpose to align axes
feature_names = X.columns  # Feature names
biplot(X_pca, pca_components, labels=feature_names)
```
